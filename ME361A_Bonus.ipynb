```python
import matplotlib.pyplot as plt
from scipy import stats

x = [0.025, 0.050, 0.100, 0.150, 0.200]
y = [-69.4665, -88.1012, -105.6237, -118.1107, -130.6807]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept
print("slope =",slope)
print("intercept =",intercept)
mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.xlabel('Feed Rate(mm/th)')
plt.ylabel('Fx(N)')
plt.show()
```
